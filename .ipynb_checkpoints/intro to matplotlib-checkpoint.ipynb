# Matplot lib example workflow

```python
# 0. import
%matplotlib inline
import matplotlib.pyplot as plt


# 1. Prepare Data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. Plot Data
ax.plot(x,y)

# 4. Customize Plot
ax.set(title = "Simple Plot",
      xlabel = "x-axis", ylabel = "y-axis")

# 5. Save and show
fig.savefig("simple-plot.png")
```

## Making figures with NumPy Arrays

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x ** 2)
```

```python
# Usafe same data and plot the data and create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from Dictionary

nut_butter_prices = { "Almond butter": 10,
                    "Peanut butter" : 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax.set(title="JK's Nut Butter Store", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="JK's Nut Butter Store", ylabel="Butter Type", xlabel="Price ($)");
```

```python
# Make some data for histogram
x = np.random.randn(1000) # Gives you normal distribution
fig, ax = plt.subplots()

ax.hist(x);
```

```python
x[:10]
```

### Options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# subplot option2 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

## Pyplot vs OO method

* Pyplot for simple ones
* Adanced ones use OO 

```python
over_50 = heart_disease[heart_disease.age > 50]
```

```python
len(over_50)
```

```python
over_50
```

```python
over_50.plot(kind="scatter", x= 'age', y='chol', c='target' );
```

```python
## OO Method with py
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax =ax);
#ax.set_xlim([0, 100])
```

```python
## OO from scratch
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y =over_50["chol"],
                    c=over_50["target"])
ax.set(title="Heart Disease and Chol Levels",
      xlabel = "Age",
      ylabel="Cholestrol Level")


ax.legend(*scatter.legend_elements(), title="Target")

# Add a headline
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols=1,
                              figsize=(10,10))

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax0.set(title = "Heart Disease and Cholesterol",
       xlabel = "Age",
       ylabel = "Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle = "--");


scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

ax1.set(title = "Heart Disease and Max heart rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["thalach"].mean(), linestyle = "--");
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols=1,
                              figsize=(10,10), sharex=True)

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax0.set(title = "Heart Disease and Cholesterol",
       ylabel = "Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle = "--");


scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

ax1.set(title = "Heart Disease and Max heart rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["thalach"].mean(), linestyle = "--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-whitegrid')

# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols=1,
                              figsize=(10,10), sharex=True)

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

ax0.set(title = "Heart Disease and Cholesterol",
       ylabel = "Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle = "--");
ax0.set_xlim([50,80])
ax0.set_ylim([100,600])

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"], cmap='winter')

ax1.set(title = "Heart Disease and Max heart rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate")

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

ax0.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle = "--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

fig.savefig("heart-disease-analysis-plot-saved-with-code.png")
```
